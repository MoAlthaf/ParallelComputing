```python
import matplotlib.pyplot as plt

# Data
explorer_labels = ['Original Explorer', 'Enhanced (BFS)']
moves = [1279, 128]
time = [0.0001, 0.0016]  # seconds

# Chart 1: Moves Made
plt.figure(figsize=(8, 5))
plt.bar(explorer_labels, moves, color='#1f77b4')
plt.title('Comparison of Moves Made by Explorers', fontsize=14, fontweight='bold')
plt.ylabel('Number of Moves', fontsize=12)
plt.xlabel('Explorer Type', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('comparison_moves_q5.png', dpi=300)
plt.show()

# Chart 2: Time Taken
plt.figure(figsize=(8, 5))
plt.bar(explorer_labels, time, color='#2ca02c')
plt.title('Comparison of Time Taken by Explorers', fontsize=14, fontweight='bold')
plt.ylabel('Time Taken (seconds)', fontsize=12)
plt.xlabel('Explorer Type', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('comparison_time_q5.png', dpi=300)
plt.show()
```
